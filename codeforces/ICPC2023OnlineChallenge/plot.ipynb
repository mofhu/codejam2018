```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_raw(fname):
    # from data to pandas dataframes

    f = open(fname,'r')
    n = int(f.readline())
    k = int(f.readline())
    t = int(f.readline())
    r = int(f.readline())
    sinr = []
    for i in range(r*k*t):
        line = [float(s) for s in f.readline().strip().split()]
        sinr.append(line)
    d = []

    for i in range(n*r*k):
        line = [float(s) for s in f.readline().strip().split()]
        d.append(line)
    j = int(f.readline())
    frames = []
    for i in range(j):
        line = [int(s) for s in f.readline().strip().split()]
        frames.append(line)
    # print(sinr)
    # print(d)
    # print(frames)

    df_sinr = pd.DataFrame(sinr)
    df_d = pd.DataFrame(d)
    df_frames = pd.DataFrame(frames)

    import seaborn as sns

    print(f'For input file {fname}')
    print(f'SINR plot:')
    sns.heatmap(df_sinr)
    plt.show()
    print(f'd plot:')
    sns.heatmap(df_d)
    plt.show()
    for i in range(1,5):
        sns.heatmap(df_frames.iloc[:, [2,i]])
        print(f'frame plot of user and column_{i}:')
        plt.show()
```

```python
plot_raw('06')
```

```python
plot_raw('21')
```
